# Wine Analysis report for both data sets(red_wine and white_wine)

##### Fixed Acidity: Acidity that doesnt evaporate. Ideal Range [ red_wine(3.0-5.5 g/L) , white_wine(5.5-7.5 g/L)]
##### Volatile Acidity: Acidity that evaporates; affects aroma. Ideal Range [ red_wine(0.2-0.6 g/L) , white_wine(0.2-0.4 g/L)]
##### Citric Acid: This can give wines a fresher taste.  Ideal Range [ red_wine(0-0.3 g/L) , white_wine(0.3-0.7 g/L)]
##### Residual Sugar: Remaining sugar after fermentation; affects sweetness. Ideal Range [ red_wine(0-2 g/L) , white_wine(1-6 g/L)]
##### Chlorides: Saltiness level in wine. Ideal Range [ red_wine(0.01-0.08 g/L) , white_wine(0.01 - 0.06 g/L)]
##### Free Sulfur Dioxide & Total Sulfur Dioxide: Preservatives used to prevent spoilage.FSD[ red_wine(10-50 mg/L) , white_wine(20-50 mg/L)] and TSD [ red_wine(20-100 mg/L) , white_wine(30-150 mg/L)]
##### Density: Linked to sugar and alcohol levels. Ideal Range [ red_wine(0.990-1.010 g/mL) , white_wine(0.990-1000 g/mL)]
##### pH: Acidity level; lower pH means higher acidity. Ideal Range [ red_wine(3.3-3.7 ) , white_wine(3.0-3.4)]
##### Sulfates: These can add to the wines stability and flavor. Ideal Range [ red_wine(0.3-0.6 g/L) , white_wine(0.3-0.7 g/L)]
##### Alcohol: Alcohol percentage. Ideal Range [ red_wine(12% - 14% ) , white_wine(10% - 13%)]
##### Quality: Rating of wine quality (usually from 0-10). Ideal Range [ red_wine(7-10 ) , white_wine(7-10)]

```python
# concat two data sets using concat operation 
col_alis = {
    'fixed acidity':'FAC', 'volatile acidity':'VAC', 'citric acid':'CAC', 'residual sugar':'RS',
       'chlorides':'CH', 'free sulfur dioxide':'FSD', 'total sulfur dioxide':'TSD', 'density':'DS',
       'pH':'PH', 'sulphates':'SLP', 'alcohol':'ALC', 'quality':'Q', 'quality_label':'QL'
}
red_df = red_wine.copy()
white_df = white_wine.copy()

red_df.rename(columns= col_alis , inplace= True)
white_df.rename(columns=col_alis, inplace= True)

red_d = red_df.describe().round(2)
white_d = white_df.describe().round(2)
pd.concat([red_d  ,white_d], keys=['red wine', 'white wine'])


```
